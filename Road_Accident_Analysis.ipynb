```python
import pandas as pd
import random

locations = ['Hyderabad', 'Vijayawada', 'Visakhapatnam', 'Warangal', 'Guntur',
             'Tirupati', 'Kurnool', 'Nellore', 'Rajahmundry', 'Karimnagar']

vehicles = ['Car', 'Bike', 'Bus', 'Truck']

weather = ['Clear', 'Rainy', 'Foggy', 'Cloudy']

data = []

for i in range(1, 501):
    data.append([
        i,
        f"2025-{random.randint(1,12):02d}-{random.randint(1,28):02d}",
        f"{random.randint(0,23):02d}:{random.randint(0,59):02d}",
        random.choice(locations),
        random.choice(vehicles),
        random.randint(0, 5),
        random.choice(weather)
    ])

df = pd.DataFrame(data, columns=[
    'Accident_ID', 'Date', 'Time', 'Location',
    'Vehicle_Type', 'Casualties', 'Weather'
])

df.to_csv('road_accident_500.csv', index=False)

print("500 records CSV file created successfully!")
df.head()
```

```python
df['Vehicle_Type'].value_counts().plot(kind='bar', figsize=(8,5))
plt.title('Accidents by Vehicle Type')
plt.show()
```

```python
df['Weather'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Weather Distribution')
plt.show()
```

```python
df['Location'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title('Accidents by Location')
plt.show()
```

```python
df['Casualties'].hist(bins=10)
plt.title('Casualties Distribution')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df.groupby(df['Date'].dt.month).size().plot(kind='line', marker='o')
plt.title('Monthly Accident Trend')
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x='Vehicle_Type', y='Casualties', data=df)
plt.title('Casualties by Vehicle Type')
plt.show()
```

```python
sns.barplot(x='Weather', y='Casualties', data=df)
plt.title('Weather vs Casualties')
plt.show()
```

```python
top_locations = df['Location'].value_counts().head(10)

top_locations.plot(kind='barh')
plt.title('Top Accident-Prone Locations')
plt.show()
```

```python
df['Vehicle_Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Vehicle Type Percentage')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64'])

sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
